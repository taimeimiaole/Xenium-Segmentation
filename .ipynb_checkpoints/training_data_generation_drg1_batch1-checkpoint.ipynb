```python
#Helper Functions
def on_segment(px, py, qx, qy, rx, ry):
    """Check if point q lies on segment pr"""
    if (qx <= max(px, rx) and qx >= min(px, rx) and
            qy <= max(py, ry) and qy >= min(py, ry)):
        return True
    return False

def orientation(px, py, qx, qy, rx, ry):
    """Find the orientation of the ordered triplet (p, q, r)
    0 -> p, q and r are collinear
    1 -> Clockwise
    2 -> Counterclockwise"""
    val = (qy - py) * (rx - qx) - (qx - px) * (ry - qy)
    if val == 0:
        return 0
    elif val > 0:
        return 1
    else:
        return 2

def do_intersect(p1x, p1y, p2x, p2y, q1x, q1y, q2x, q2y):
    """Check if line segment 'p1q1' and 'p2q2' intersect"""
    # Find the four orientations needed for general and special cases
    o1 = orientation(p1x, p1y, q1x, q1y, p2x, p2y)
    o2 = orientation(p1x, p1y, q1x, q1y, q2x, q2y)
    o3 = orientation(p2x, p2y, q2x, q2y, p1x, p1y)
    o4 = orientation(p2x, p2y, q2x, q2y, q1x, q1y)

    # General case
    if o1 != o2 and o3 != o4:
        return True

    # Special cases
    # p1, q1 and p2 are collinear and p2 lies on segment p1q1
    if o1 == 0 and on_segment(p1x, p1y, p2x, p2y, q1x, q1y):
        return True

    # p1, q1 and q2 are collinear and q2 lies on segment p1q1
    if o2 == 0 and on_segment(p1x, p1y, q2x, q2y, q1x, q1y):
        return True

    # p2, q2 and p1 are collinear and p1 lies on segment p2q2
    if o3 == 0 and on_segment(p2x, p2y, p1x, p1y, q2x, q2y):
        return True

    # p2, q2 and q1 are collinear and q1 lies on segment p2q2
    if o4 == 0 and on_segment(p2x, p2y, q1x, q1y, q2x, q2y):
        return True

    # If none of the cases
    return False
```

```python
#Read in and Scale Manual Annotations
df = pd.read_csv('/media/yhs/5596744f-db7c-442f-9235-d0c9d50c0a6b/export/manual_annotations/cellids1.csv')
l1 = list(df.cell_id.unique())
l2 = [] 
for i in range(len(l1)):
    dict1 = {l1[i]:[]}
    l2.append(dict1)
df['y']*=1/4.707
df['x']*=1/4.707
```

```python
#Process the annotaions
j=0
for i in range(len(l2)):
    id = list(l2[i].keys())[0]
    while j< (df.shape[0]):
        print(df.loc[j, :]['cell_id'])
        if df.loc[j, :]['cell_id']==id:
            #can cast to int below if needed
            x = (int)(df.loc[j, :]['x'])
            y = (int)(df.loc[j, :]['y'])
            l2[i][id].append((x,y))
            j+=1
        else:
            break
```

```python
#Read in the corresp. annotated image to get a respsective size boundary
ST = cv2.imread('/media/yhs/5596744f-db7c-442f-9235-d0c9d50c0a6b/export/transcripts/output-XETG00171__0018220__Region_1__20240207__003742.tif')
ST = cv2.cvtColor(ST, cv2.COLOR_BGR2RGB)
```

```python
#Assign the annotations to specific regions
list1=[]
for i in range (0, ST.shape[0], 500):
    list1.append([(0,i), (ST.shape[1], i)])
for j in range (0, ST.shape[1], 500):
    list1.append([(j,0),(j, ST.shape[0])]) 
#Deepcopy l2 to duplicate it for processing of Csvs
l3= deepcopy(l2)
```

```python
x_intersections=0
y_intersections=0
intersections_1=0
intersected_cells=[]
#Remove intersections
linesdrawn=0
cellsdrawn=0
color=(255,255,0)
thickness= 1
for i in range(len(l2)):
    pts = list(l2[i].values())[0]
    intersect = False
    #CHECK if any points intersect the line segments 
    for a in range (len(list1)):
        if not intersect: 
            p1x = list1[a][0][0]
            p1y = list1[a][0][1]
            p2x = list1[a][1][0]
            p2y = list1[a][1][1]
            for j in range(len(pts)-1):
                p3x = pts[j][0]
                p3y = pts[j][1]
                p4x = pts[j+1][0]
                p4y = pts[j+1][1]
                if do_intersect(p1x, p1y, p3x, p3y, p2x, p2y, p4x, p4y):
                    intersect = True
                    break
    #If not then draw 
    #Else break the draw and go next 
    if not intersect:
        for j in range (len(pts)-1):
            ST= cv2.line(ST, pts[j], pts[j+1], color, thickness) 
    else:
        l3[i]=0
```

```python
def f7(seq):
    seen = set()
    seen_add = seen.add
    return [x for x in seq if not (x in seen or seen_add(x))]
#Remove Duplicates 
for i in range(len(l2)): 
    new_values = f7(list(l2[i].values())[0])
    key_rep = list(l2[i].keys())[0]
    l2[i][key_rep] = new_values
```

```python
#Add Start Point to annotations for cells
for i in range(len(l2)):
    key = list(l2[i].keys())[0]
    print(key)
    first_value = list(l2[i].values())[0][0]
    last_index = len(list(l2[i].values())[0])-1
    last_value = list(l2[i].values())[0][last_index]
    if first_value != last_value:
        new_list1 = list(l2[i].values())[0]
        new_list1.append(first_value)
        l2[i][key] = new_list1
#filter out 0s after running the drawing
l3[:] = (value for value in l3 if value != 0)
```

```python
#Float Coord
#[0][0] is x [0][1] is y
# i is y, j is x
#All empty Csvs are removed
csv=0
for i in range (0, STClean.shape[0]-500, 500):
    for j in range (0, STClean.shape[1]-500, 500):
        dict1 ={}
        total_x=[]
        total_y=[]
        total_names=[]
        for a in range(len(l3)):
            if (list(l3[a].values())[0][0][0]>j and list(l3[a].values())[0][0][0]<j+500 and list(l3[a].values())[0][0][1]>i and list(l3[a].values())[0][0][1]<i+500):
                name = list(l3[a].keys())[0]
                name_list = []
                x_cord = [i[0] for i in list(l3[a].values())[0]]
                x_cord = ["%.6f" % round((x-j)/500, 6) for x in x_cord]
                y_cord = [i[1] for i in list(l3[a].values())[0]]
                y_cord = ["%.6f" % round((y-i)/500, 6) for y in y_cord]
                name_list += len(x_cord) * [name]
                total_x+=x_cord
                total_y+=y_cord
                total_names+=name_list
        dict1['cell_id'] = total_names
        dict1['x'] = total_x
        dict1['y'] = total_y
        new = pd.DataFrame.from_dict(dict1)
        print(i,j)
        print(new)
        new.to_csv(csv_path+f"cropped_csv_{str(i)+'_'+str(j)}.csv", index=False)
```

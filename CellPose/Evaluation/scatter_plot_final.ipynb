```python
import os
import re
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import ast
import numpy as np

def scatter_creation(model, batch, e, folder):
    data = {}
    ep = []
    f1_50 = []
    f1_5095 = []
    for file in os.listdir(folder):
        if model + '_' in file and ('RN' + batch in file or 'RB' + batch in file) and e in file:
            print(file)
            num = re.search(r'_(\d+)epochs', file)
            ep.append(num.group(1))
            df = pd.read_csv(os.path.join(folder, file))
            m50 = df['cell_f1_50'].tolist()[-1]
            m5095 = df['cell_f1_50-95'].tolist()[-1]
            f1_50.append(m50)
            f1_5095.append(m5095)
    data['Epochs'] = ep
    data['cell_f1_50'] = f1_50
    data['cell_f1_50-95'] = f1_5095
    return data

name = 'mergedDRG123'
batch = '2'
ev_imgs = 'drg4'
e_dir = '/media/yhs/5596744f-db7c-442f-9235-d0c9d50c0a6b/Cellpose/DRG123_Merged_New/Results/Epochs/'
data = scatter_creation(name, batch, ev_imgs, e_dir)

sorted_indices = sorted(range(len(data['Epochs'])), key=lambda i: int(data['Epochs'][i]))
ordered_data = {key: [values[i] for i in sorted_indices] for key, values in data.items()}

df = pd.DataFrame(ordered_data)
df.to_csv(os.path.join(e_dir, 'categories.csv'), index=False)

max_F1_50 = max(data['cell_f1_50'])
max_F1_50_95 = max(data['cell_f1_50-95'])
max_index_F1_50 = ordered_data['cell_f1_50'].index(max_F1_50)
max_index_F1_50_95 = ordered_data['cell_f1_50-95'].index(max_F1_50_95)

print(f"Max F1_50 occurs at Epochs: {ordered_data['Epochs'][max_index_F1_50]}")
print(f"Max F1_50-95 occurs at Epochs: {ordered_data['Epochs'][max_index_F1_50_95]}")

df = pd.DataFrame(ordered_data)
df.to_csv(os.path.join(e_dir, name + 'scatter.csv'), index=False)

# Set Seaborn style and context
sns.set(style="whitegrid")
sns.set_context("talk", font_scale=2)  # Adjust context and font scale for larger proportions

# Create scatter plot with lines
plt.figure(figsize=(14, 8))  # Increase figure size
sns.lineplot(x='Epochs', y='cell_f1_50', data=df, marker='o', label='F1_50', color='green')
sns.lineplot(x='Epochs', y='cell_f1_50-95', data=df, marker='x', label='F1_50-95', color='red')

# Add labels and title with bold and appealing fonts
plt.xlabel('Epochs', fontsize=16, weight='bold', family='serif')  # Bold x-axis label
plt.ylabel('f1-score', fontsize=16, weight='bold', family='serif')  # Bold y-axis label
plt.xticks(fontsize=14, family='serif')  # Larger font size for x-axis values
plt.yticks(fontsize=14, family='serif')  # Larger font size for y-axis values
plt.title(name + ' Epochs vs F1_50 and F1_50-95 run on ' + ev_imgs, 
          fontsize=20, weight='bold', family='serif')  # Bold and larger title

# Add a legend with larger font size and serif font
plt.legend(fontsize=14, title='Legend', title_fontsize=16, frameon=True)

# Save plot as .svg image
plt.savefig(name + "_vs_f1-score_" + ev_imgs + ".svg", format='svg', dpi=300)

# Show the plot
plt.show()

```

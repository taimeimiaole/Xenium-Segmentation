```python
#INPUTS: YOLO TXT file/directory and size of image to map the masks onto
#OUTPUTS: Custom csv to be used for downstream evaluation and visualization
import pandas as pd
import os
import re

def extracting_rois(file_path, width, height):
    with open(file_path, 'r') as file:
        lines = file.read().splitlines()
    rois = []
    for i in lines:
        rois.append(list(map(float, i.split())))
    if len(rois) > 0:
        print(len(rois[0]))
    rois_filtered = []
    for i in rois:
        rois_filtered.append(i[1:])
    if len(rois_filtered) > 0:
        print(len(rois_filtered[0]))
    int_rois = []
    for i in rois_filtered:
        count = 0
        x = []
        y = []
        for j in i:
            if count%2:
                val = int(round(j*width))
                x.append(val)
            else:
                val = int(round(j*height))
                y.append(val)
            count += 1
        temp = list(zip(x, y))
        filt = []
        for i in temp:
            if i not in filt:
                filt.append(i)
        int_rois.append(filt)
    return int_rois

#Only if files in folder are copies of something (aligns the prediction file names with the ground truth file names)
def standardizing_filenames(directory):
    # Loop through all files in the directory
    for filename in os.listdir(directory):
        old_file = os.path.join(directory, filename)
        if os.path.isdir(old_file):
            standardizing_filenames(old_file)
        else:
            # Use regex to find files with ' (1)' in the filename
            new_filename = re.sub(r' \(\d+\)', '', filename)
            
            # Construct full file paths
            
            new_file = os.path.join(directory, new_filename)
            
            # Rename the file if the filename was changed
            if new_filename != filename:
                os.rename(old_file, new_file)
                print(f'Renamed: {old_file} -> {new_file}')

#Recursive call to extract_rois through a folder
def extract_rec(directory, width, height):
    # Loop through all files in the directory
    for filename in os.listdir(directory):
        file = os.path.join(directory, filename)
        if os.path.isdir(file):
            extract_rec(file, width, height)
        else:
            if '.txt' in filename:
                data = extracting_rois(file, width, height)
                df = {'Name': [], 'X': [], 'Y': []}
                count = 1
                for i in data:
                    df['Name'].extend(['roi'+str(count) for j in i])
                    x_list = [k for j, k in i]
                    y_list = [j for j, k in i]
                    df['X'].extend(x_list)
                    df['Y'].extend(y_list)
                    count += 1
                df_csv = pd.DataFrame(df)
                out_dir = os.path.join(directory, 'csv')
                os.makedirs(out_dir, exist_ok=True)
                df_csv.to_csv(os.path.join(out_dir, filename.replace('.txt','.csv')))


standardizing_filenames(directory)
extract_rec(directory, w, h)
```

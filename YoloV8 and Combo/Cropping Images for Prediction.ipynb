```python
#cropping the DRG1 Merged Image for Predition (DRG1 only had to be cropped for Merged dataset)

from PIL import Image
import os

def crop_image_with_custom_coords(image_path, crop_coordinates):
    """
    Crops an image based on user-specified coordinates.
    
    :param image_path: Path to the image.
    :param crop_coordinates: List of tuples specifying crop coordinates (left, upper, right, lower).
    :return: List of cropped image data with details.
    """
    # Open the image
    image = Image.open(image_path)
    
    cropped_images = []
    for i, (left, upper, right, lower) in enumerate(crop_coordinates):
        cropped_image = image.crop((left, upper, right, lower))
        cropped_images.append({
            "crop_index": i + 1,
            "image": cropped_image,
            "coordinates": (left, upper, right, lower),
            "dimensions": cropped_image.size,
        })

    return cropped_images

def save_cropped_images(cropped_images, output_folder):
    """
    Saves cropped images to the specified output folder.
    
    :param cropped_images: List of cropped image data with details.
    :param output_folder: Directory where cropped images will be saved.
    """
    # Ensure the output directory exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    for crop_data in cropped_images:
        index = crop_data["crop_index"]
        cropped_image = crop_data["image"]
        filename = f"{output_folder}/crop_{index}.tiff"
        cropped_image.save(filename)
        print(f"Crop {index}: Saved as {filename}")
        print(f"  Coordinates: {crop_data['coordinates']}")
        print(f"  Dimensions: {crop_data['dimensions']}")

if __name__ == "__main__":
    # Example inputs
    image_path = r"path to full original input image"  # Replace with the path to your image
    output_directory = r"path to output for cropped images"  # Replace with your desired output directory
    
    # Define custom crop coordinates (left, upper, right, lower) in pixels
    crop_coordinates = [
        (0, 0, 2400, 5797),     # First crop
        (2200, 0, 7261, 5797), # Second crop
    ]
    
    # Perform cropping and save images
    cropped_images = crop_image_with_custom_coords(image_path, crop_coordinates)
    save_cropped_images(cropped_images, output_directory)
```
